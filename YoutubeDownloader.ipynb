```
import os
import sys
import time
import subprocess
import re
import urllib.request
import platform
import stat
from IPython.display import display, HTML, clear_output

def log_ui(message, percent):
    html = f"""
    <div style='font-family: "Inter", sans-serif; background: #0f172a; padding: 25px; border-radius: 16px; color: white; width: 100%; max-width: 500px; margin: 0 auto; box-shadow: 0 10px 25px rgba(0,0,0,0.5);'>
        <h3 style='margin: 0 0 15px 0; color: #38bdf8; font-weight: 800; font-size: 20px;'>⚡ Initializing Engine...</h3>
        <div style='background: rgba(255,255,255,0.1); border-radius: 10px; height: 10px; overflow: hidden; margin-bottom: 15px;'>
            <div style='background: linear-gradient(90deg, #38bdf8, #818cf8); width: {percent}%; height: 100%; transition: width 0.5s ease;'></div>
        </div>
        <p style='margin: 0; font-size: 14px; font-weight: 600; color: #cbd5e1; display: flex; align-items: center; gap: 10px;'>
            <div style='width: 16px; height: 16px; border: 3px solid #38bdf8; border-top-color: transparent; border-radius: 50%; animation: spin 1s linear infinite;'></div>
            {message}
        </p>
        <style>@keyframes spin {{ 100% {{ transform: rotate(360deg); }} }}</style>
    </div>
    """
    clear_output(wait=True)
    display(HTML(html))

log_ui("Purging old orphan processes...", 10)
subprocess.run(["pkill", "-f", "uvicorn"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
subprocess.run(["pkill", "-f", "cloudflared"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
subprocess.run(["pkill", "-f", "ssh"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

system = platform.system().lower()
machine = platform.machine().lower()

log_ui("Syncing with GitHub for Auto-Healing...", 30)
if os.path.exists("YoutubeDownloader/.git"):
    os.chdir("YoutubeDownloader")
    subprocess.run(["git", "reset", "--hard"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    subprocess.run(["git", "pull", "origin", "main"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
elif not os.path.exists("main.py"):
    subprocess.run(["git", "clone", "https://github.com/RashidKhanApDev/YoutubeDownloader.git"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    if os.path.exists("YoutubeDownloader"):
        os.chdir("YoutubeDownloader")
elif os.path.exists("YoutubeDownloader"):
    os.chdir("YoutubeDownloader")

log_ui("Installing advanced dependencies...", 50)
if os.path.exists("requirements.txt"):
    subprocess.run([sys.executable, "-m", "pip", "install", "-r", "requirements.txt", "--quiet"])

log_ui("Igniting FastAPI Server...", 70)
subprocess.Popen([sys.executable, "-m", "uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(3)

log_ui("Establishing Global Secure Tunnel...", 85)
url = ""

# 1. Primary Tunnel: Cloudflare
cf_binary = "cloudflared"
if system == "windows":
    cf_url = "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-windows-amd64.exe"
    cf_binary = "cloudflared.exe"
elif system == "darwin": 
    cf_url = "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-darwin-amd64.tgz"
else:
    if "arm" in machine or "aarch" in machine:
        cf_url = "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-arm64"
    else:
        cf_url = "https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64"

try:
    if not os.path.exists(cf_binary):
        if system == "darwin":
            subprocess.run(["curl", "-L", "-o", "cloudflared.tgz", cf_url], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
            subprocess.run(["tar", "-xzf", "cloudflared.tgz"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        else:
            urllib.request.urlretrieve(cf_url, cf_binary)
        if system != "windows" and os.path.exists(cf_binary):
            st = os.stat(cf_binary)
            os.chmod(cf_binary, st.st_mode | stat.S_IEXEC)
except:
    pass

try:
    if os.path.exists("cloudflared.log"):
        os.remove("cloudflared.log")
    cf_cmd = f"./{cf_binary}" if system != "windows" else cf_binary
    subprocess.Popen([cf_cmd, "tunnel", "--url", "http://localhost:8000", "--logfile", "cloudflared.log"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    
    for _ in range(15):
        if os.path.exists("cloudflared.log"):
            with open("cloudflared.log", "r", encoding="utf-8") as f:
                match = re.search(r"https://[a-zA-Z0-9-]+\\.trycloudflare\\.com", f.read())
                if match:
                    url = match.group(0)
                    break
        time.sleep(1)
except:
    pass

# 2. Secondary Tunnel: Pinggy SSH (Absolute Fallback)
if not url and system != "windows":
    log_ui("Cloudflare timeout. Trying Secondary SSH Tunnel...", 90)
    try:
        if os.path.exists("pinggy.log"):
            os.remove("pinggy.log")
        subprocess.Popen(["ssh", "-p", "443", "-R0:localhost:8000", "-o", "StrictHostKeyChecking=no", "-o", "ServerAliveInterval=30", "a.pinggy.io"], stdout=open("pinggy.log", "w"), stderr=subprocess.STDOUT)
        for _ in range(10):
            if os.path.exists("pinggy.log"):
                with open("pinggy.log", "r", encoding="utf-8") as f:
                    match = re.search(r"https://[a-zA-Z0-9-]+\\.a\\.pinggy\\.link", f.read())
                    if match:
                        url = match.group(0)
                        break
            time.sleep(1)
    except:
        pass

if not url:
    url = "http://127.0.0.1:8000"
    status_text = "Local Private Server Online"
    desc_text = "Your high-speed server is running locally on your device."
    try:
        import webbrowser
        webbrowser.open(url)
    except:
        pass
else:
    status_text = "Global Cloud System Online"
    desc_text = "Your high-speed private server is securely tunneled with Advanced Fallbacks."

clear_output()

html_code = f"""
<style>
@import url('https://fonts.googleapis.com/css2?family=Inter:wght@400;600;800&display=swap');

.app-container {{
    display: flex;
    justify-content: center;
    align-items: center;
    padding: 40px 20px;
    font-family: 'Inter', sans-serif;
    background-color: transparent;
}}

.premium-card {{
    background: linear-gradient(145deg, #0f9b4f 0%, #0a6c36 100%);
    border: 1px solid rgba(255,255,255,0.15);
    padding: 45px 40px;
    border-radius: 24px;
    box-shadow: 0 25px 50px rgba(10, 108, 54, 0.3), inset 0 1px 0 rgba(255,255,255,0.2);
    text-align: center;
    color: white;
    max-width: 500px;
    width: 100%;
    animation: slideUp 0.6s cubic-bezier(0.16, 1, 0.3, 1);
    position: relative;
    overflow: hidden;
}}

.premium-card::before {{
    content: '';
    position: absolute;
    top: -50%; left: -50%; width: 200%; height: 200%;
    background: radial-gradient(circle, rgba(255,255,255,0.08) 0%, transparent 60%);
    transform: rotate(30deg);
    pointer-events: none;
}}

@keyframes slideUp {{
    0% {{ transform: translateY(30px) scale(0.95); opacity: 0; }}
    100% {{ transform: translateY(0) scale(1); opacity: 1; }}
}}

.status-badge {{
    display: inline-flex;
    align-items: center;
    background: rgba(255, 255, 255, 0.15);
    padding: 8px 18px;
    border-radius: 50px;
    font-size: 13px;
    font-weight: 600;
    letter-spacing: 0.5px;
    margin-bottom: 25px;
    backdrop-filter: blur(10px);
    border: 1px solid rgba(255,255,255,0.1);
}}

.status-dot {{
    width: 8px; height: 8px;
    background-color: #4ade80;
    border-radius: 50%;
    margin-right: 10px;
    box-shadow: 0 0 10px #4ade80;
    animation: pulse 2s infinite;
}}

@keyframes pulse {{
    0% {{ box-shadow: 0 0 0 0 rgba(74, 222, 128, 0.4); }}
    70% {{ box-shadow: 0 0 0 10px rgba(74, 222, 128, 0); }}
    100% {{ box-shadow: 0 0 0 0 rgba(74, 222, 128, 0); }}
}}

.title {{
    margin: 0 0 12px 0;
    font-weight: 800;
    font-size: 30px;
    letter-spacing: -0.5px;
}}

.subtitle {{
    font-size: 16px;
    opacity: 0.9;
    margin-bottom: 35px;
    line-height: 1.5;
}}

.link-box {{
    background: rgba(0, 0, 0, 0.25);
    border: 1px solid rgba(255, 255, 255, 0.1);
    border-radius: 14px;
    padding: 20px;
    margin-bottom: 30px;
    position: relative;
    transition: all 0.3s ease;
}}

.link-box:hover {{
    background: rgba(0, 0, 0, 0.35);
    border-color: rgba(255, 255, 255, 0.3);
}}

.link-label {{
    display: block;
    text-align: left;
    font-size: 12px;
    text-transform: uppercase;
    letter-spacing: 1.5px;
    font-weight: 600;
    color: #a7f3d0;
    margin-bottom: 10px;
}}

.link-input {{
    width: 100%;
    background: transparent;
    border: none;
    color: white;
    font-size: 17px;
    font-family: 'Inter', monospace;
    font-weight: 600;
    text-align: left;
    outline: none;
    cursor: text;
}}

.copy-button {{
    background: white;
    color: #0a6c36;
    border: none;
    padding: 16px 32px;
    border-radius: 50px;
    font-size: 17px;
    font-weight: 800;
    cursor: pointer;
    width: 100%;
    display: flex;
    align-items: center;
    justify-content: center;
    gap: 12px;
    box-shadow: 0 8px 20px rgba(0,0,0,0.15);
    transition: all 0.3s cubic-bezier(0.16, 1, 0.3, 1);
}}

.copy-button:hover {{
    transform: translateY(-4px);
    box-shadow: 0 12px 25px rgba(0,0,0,0.25);
    background: #f8fafc;
}}

.copy-button:active {{
    transform: translateY(0);
}}

.footer-text {{
    margin-top: 30px;
    font-size: 13px;
    opacity: 0.6;
    font-weight: 600;
}}
</style>

<div class="app-container">
    <div class="premium-card">
        <div class="status-badge">
            <div class="status-dot"></div>
            {status_text}
        </div>
        <h2 class="title">Turbo Downloader</h2>
        <p class="subtitle">{desc_text}</p>
        
        <div class="link-box">
            <span class="link-label">Secure Access Link</span>
            <input type="text" class="link-input" id="linkText" value="{url}" readonly onclick="this.select()">
        </div>
        
        <button class="copy-button" onclick="copyDetails()" id="actionBtn">
            <svg width="22" height="22" fill="none" stroke="currentColor" stroke-width="2.5" viewBox="0 0 24 24"><path d="M8 16H6a2 2 0 01-2-2V6a2 2 0 012-2h8a2 2 0 012 2v2m-6 12h8a2 2 0 002-2v-8a2 2 0 00-2-2h-8a2 2 0 00-2 2v8a2 2 0 002 2z"></path></svg>
            Open / Copy Access Link
        </button>
        
        <div class="footer-text">Powered by Advanced Fallback Architecture</div>
    </div>
</div>

<script>
function playPremiumChime() {{
    try {{
        const AudioContext = window.AudioContext || window.webkitAudioContext;
        const ctx = new AudioContext();
        function playTone(freq, time, dur) {{
            const osc = ctx.createOscillator();
            const gain = ctx.createGain();
            osc.type = 'sine';
            osc.frequency.setValueAtTime(freq, ctx.currentTime + time);
            gain.gain.setValueAtTime(0, ctx.currentTime + time);
            gain.gain.linearRampToValueAtTime(0.15, ctx.currentTime + time + 0.05);
            gain.gain.exponentialRampToValueAtTime(0.001, ctx.currentTime + time + dur);
            osc.connect(gain);
            gain.connect(ctx.destination);
            osc.start(ctx.currentTime + time);
            osc.stop(ctx.currentTime + time + dur);
        }}
        playTone(523.25, 0, 0.6);   // C5
        playTone(659.25, 0.1, 0.8); // E5
        playTone(1046.50, 0.2, 1.5); // C6
    }} catch(e) {{ console.log('Audio API not supported'); }}
}}

playPremiumChime();

function copyDetails() {{
    var link = document.getElementById("linkText").value;
    var btn = document.getElementById("actionBtn");
    
    if (link.includes("127.0.0.1") || link.includes("localhost") || link.includes(".com") || link.includes(".link")) {{
        window.open(link, "_blank");
    }}
    
    var originalHTML = btn.innerHTML;
    var successHTML = '<svg width="22" height="22" fill="none" stroke="currentColor" stroke-width="2.5" viewBox="0 0 24 24"><path stroke-linecap="round" stroke-linejoin="round" d="M5 13l4 4L19 7"></path></svg> Link Copied!';
    
    if(navigator.clipboard && window.isSecureContext) {{
        navigator.clipboard.writeText(link).then(function() {{
            btn.innerHTML = successHTML;
            btn.style.color = "#0a6c36";
            setTimeout(function() {{ btn.innerHTML = originalHTML; }}, 2500);
        }});
    }} else {{
        var dummy = document.createElement("textarea");
        document.body.appendChild(dummy);
        dummy.value = link;
        dummy.select();
        document.execCommand("copy");
        document.body.removeChild(dummy);
        btn.innerHTML = successHTML;
        setTimeout(function() {{ btn.innerHTML = originalHTML; }}, 2500);
    }}
}}
</script>
"""
display(HTML(html_code))

try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    pass
```
