```
!git clone https://github.com/RashidKhanApDev/YoutubeDownloader.git > /dev/null 2>&1
%cd YoutubeDownloader
!pip install -r requirements.txt > /dev/null 2>&1
!npm install -g localtunnel > /dev/null 2>&1

import subprocess
import time
import urllib.request
from IPython.display import display, HTML

print("Starting Server... Please wait a moment...")
subprocess.Popen(["python", "-m", "uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(3)

print("Generating Public Link...")
lt_process = subprocess.Popen(["npx", "localtunnel", "--port", "8000"], stdout=subprocess.PIPE, text=True)

url = ""
for line in lt_process.stdout:
    if "your url is" in line.lower():
        url = line.split("is")[-1].strip()
        break

ip = urllib.request.urlopen("https://ipv4.icanhazip.com").read().decode("utf8").strip()

html_code = f"""
<style>
.custom-modal-overlay {{
    position: fixed;
    top: 0; left: 0; right: 0; bottom: 0;
    background: rgba(0, 0, 0, 0.6);
    display: flex;
    align-items: center;
    justify-content: center;
    z-index: 9999;
    backdrop-filter: blur(5px);
}}
.custom-modal {{
    background: linear-gradient(135deg, #0ba360 0%, #3cba92 100%);
    padding: 30px;
    border-radius: 20px;
    box-shadow: 0 15px 35px rgba(0,0,0,0.3);
    text-align: center;
    color: white;
    font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    max-width: 450px;
    width: 90%;
    animation: popIn 0.5s cubic-bezier(0.175, 0.885, 0.32, 1.275);
}}
@keyframes popIn {{
    0% {{ transform: scale(0.8); opacity: 0; }}
    100% {{ transform: scale(1); opacity: 1; }}
}}
.custom-modal h2 {{
    margin-top: 0; color: white; font-weight: 800; letter-spacing: 1px;
}}
.custom-input-group {{
    margin: 20px 0;
    background: rgba(255,255,255,0.2);
    padding: 15px;
    border-radius: 12px;
    border: 1px solid rgba(255,255,255,0.3);
}}
.custom-input-group p {{
    margin: 0 0 8px 0; font-size: 14px; font-weight: 600; opacity: 0.9;
}}
.custom-input-group input {{
    width: 90%;
    padding: 10px;
    border: none;
    border-radius: 8px;
    text-align: center;
    font-weight: bold;
    color: #0ba360;
    background: white;
    font-size: 16px;
}}
.copy-btn {{
    background: #ffffff;
    color: #0ba360;
    border: none;
    padding: 12px 25px;
    border-radius: 50px;
    font-weight: bold;
    font-size: 16px;
    cursor: pointer;
    margin-top: 10px;
    transition: all 0.3s ease;
    box-shadow: 0 4px 15px rgba(0,0,0,0.1);
}}
.copy-btn:hover {{
    transform: translateY(-3px);
    box-shadow: 0 6px 20px rgba(0,0,0,0.2);
}}
.close-btn {{
    margin-top: 15px;
    cursor: pointer;
    font-size: 14px;
    opacity: 0.8;
    text-decoration: underline;
}}
.close-btn:hover {{ opacity: 1; }}
</style>

<div class="custom-modal-overlay" id="myModal">
    <div class="custom-modal">
        <h2>✅ SERVER IS READY</h2>
        <p>Your cloud downloader is running successfully!</p>
        
        <div class="custom-input-group">
            <p>🔗 PUBLIC LINK:</p>
            <input type="text" id="linkText" value="{url}" readonly>
        </div>
        
        <div class="custom-input-group">
            <p>🔑 PASSWORD (Endpoint IP):</p>
            <input type="text" id="ipText" value="{ip}" readonly>
        </div>
        
        <button class="copy-btn" onclick="copyDetails()">📋 Copy Link & Password</button>
        <div class="close-btn" onclick="document.getElementById('myModal').style.display='none'">Dismiss</div>
    </div>
</div>

<script>
function copyDetails() {{
    var link = document.getElementById("linkText").value;
    var ip = document.getElementById("ipText").value;
    var textToCopy = "👉 Link: " + link + "\n👉 Password: " + ip;
    var btn = document.querySelector(".copy-btn");
    
    navigator.clipboard.writeText(textToCopy).then(function() {{
        btn.innerHTML = "✅ Copied Successfully!";
        setTimeout(function() {{ btn.innerHTML = "📋 Copy Link & Password"; }}, 3000);
    }}).catch(function() {{
        var dummy = document.createElement("textarea");
        document.body.appendChild(dummy);
        dummy.value = textToCopy;
        dummy.select();
        document.execCommand("copy");
        document.body.removeChild(dummy);
        btn.innerHTML = "✅ Copied Successfully!";
        setTimeout(function() {{ btn.innerHTML = "📋 Copy Link & Password"; }}, 3000);
    }});
}}
</script>
"""
display(HTML(html_code))

# Keep tunnel alive
lt_process.wait()
```
